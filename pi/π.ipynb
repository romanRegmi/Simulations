```python
import time
import numpy as np
from itertools import count
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.title('Value of Pi using Monte Carlo Simulation')
plt.xlabel("Time")
plt.ylabel("Pi")
fig.show()
```

```python
circle = 0
square = 0
outside = 0
```

```python
h = -2.5
k = 0
plot_x_axis=[]
index = count()
```

```python
pi = []
for i in range(100):
    for j in range(1000):
        balls_x = np.random.uniform(-5, 3.5) # rectangle x-axis ranges from -5 to 3.5
        balls_y = np.random.uniform(-2.5,2.5) # rectange y-axis ranges from -2.5 to 2.5
        
        r2 = (balls_x-h)**2+(balls_y-k)**2
    
        if balls_x<=0 and r2<=6.25:
            circle=circle+1
        elif balls_x>=1 and balls_x<=3.5 and balls_y>=0:
            square = square+1
        else:
            outside = outside+1
    
    plot_x_axis.append(next(index))
    pi.append(circle/square)
    fig.canvas.draw()
    ax.plot(plot_x_axis, pi, color='b')
    time.sleep(0.1)
```
